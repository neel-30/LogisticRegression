```python
from sklearn.preprocessing import LabelEncoder   #changing object data type to numeric datatype
```

```python
le=LabelEncoder()
```

```python
for i in df.columns:
    if df[i].dtype=='object':
        df[i]=le.fit_transform(df[i])
```

```python
df.info()
```

```python
x=df.iloc[: ,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
rfe=RFE(model)
```

```python
rfe.fit(x,y)
```

```python
col_names=x.columns
```

```python
select_indices=[i for i ,data in enumerate(rfe.support_) if data==True]
```

```python
select_indices
```

```python
selected_cols=[col_names[i] for i in select_indices]
selected_cols
```

```python
x=x.loc[:,selected_cols]
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=.80,random_state=0)
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
ans=lr.predict(x_test)
```

```python
ans
```

```python
from sklearn.metrics import*
```

```python
accuracy_score(ans,y_test)  #90%accuracy
```

```python
confusion_matrix(ans,y_test)
```

```python
report=classification_report(ans,y_test)
print(report)
```
